```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import tensorflow as tf
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)
from tensorflow import keras
import warnings
warnings.filterwarnings('ignore', category=UserWarning)
from tensorflow.keras.losses import Huber
from sklearn.preprocessing import MinMaxScaler
```

```python
# Baca data Excel
df_train = pd.read_excel('lomba/train_prepro.xlsx', index_col=0)
df_test = pd.read_excel('lomba/test_prepro.xlsx',  index_col=0)
```

```python
df_train
```

```python
df_test
```

```python
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

# Pisahkan fitur dan target
X_train_scaled = df_train.iloc[:, :-1]
y_train_scaled = df_train.iloc[:, -1]

data_X_train_scaled = scaler_X.fit_transform(X_train_scaled)
data_y_train_scaled = scaler_y.fit_transform(np.array(y_train_scaled).reshape(-1, 1))

data_test_scaled = scaler_X.fit_transform(df_test)
```

```python
X_train_reshape = data_X_train_scaled.reshape(data_X_train_scaled.shape[0], data_X_train_scaled.shape[1], 1)
X_test_reshape = data_test_scaled.reshape(data_test_scaled.shape[0], data_test_scaled.shape[1], 1)
```

```python
from tensorflow.keras.metrics import RootMeanSquaredError
def modelnya():
    model = tf.keras.models.Sequential([
    tf.keras.layers.Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train_reshape.shape[1], 1)),
    tf.keras.layers.MaxPooling1D(pool_size=2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu', kernel_regularizer='l2'),
    tf.keras.layers.Dense(1, activation='linear')  # Aktivasi 'linear' untuk regresi
    ])
    model.compile(optimizer='sgd', loss='mean_squared_error', metrics=[RootMeanSquaredError()])
    return model
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, TensorBoard
model = modelnya()
  # Definisikan callbacks
callbacks = [
        ModelCheckpoint(f'./logs/lomba/best_model.h5', monitor='val_loss', save_best_only=True),
        TensorBoard(log_dir=f'./logs/lomba', histogram_freq=1, write_graph=True, write_images=True)
    
    ]
history = model.fit(X_train_reshape, data_y_train_scaled, validation_split=0.1, batch_size=64, epochs=20,
                        verbose=2, callbacks=callbacks, workers=-1)
```

```python
# Plot Loss dan Validation Loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Train and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Plot Loss dan Validation Loss
plt.plot(history.history['root_mean_squared_error'], label='Train rmse')
plt.plot(history.history['val_root_mean_squared_error'], label='Validation rmse')
plt.title('Train and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Prediksi menggunakan model
predictions = model.predict(X_test_reshape)
```

```python
balikan_prediksi = scaler_y.inverse_transform(predictions).astype(int)
predictions_float = balikan_prediksi.astype(float)
```
